# Sales Country Analysis

Goals: Show data what is the best product sales in each country?

```python
#import library

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read data
data=pd.read_csv('sales_each_country.csv')
data
```

```python
sns.relplot(x = 'ProductName', y = 'Sum_Quantity', col = 'Country' ,data = data.head(100),kind='line',height=8.27, aspect=17.7/8.27)
plt.show() 
```

```python
#create country data that contain country and sum quantity
country=data.groupby("Country")["Sum_Quantity"].max().sort_values(ascending= False).head(25)
country
```

```python
plt.figure(figsize=(16,10))
ax = sns.barplot(x=country, y=country.index)

plt.xlabel("Quantity")
plt.ylabel("Country");
plt.title('Which country has most number of sales?');


for patch in ax.patches:
    width = patch.get_width() 
    height = patch.get_height() 
    x = patch.get_x() 
    y = patch.get_y()
    
    plt.text(width + x, height + y, ' {:.1f} '.format(width))
```

```python
product=data.groupby("ProductName")["Sum_Quantity"].max().sort_values(ascending= False).head(25)
product
```

```python
plt.figure(figsize=(16,10))
ax = sns.barplot(x=product, y=product.index)

plt.xlabel("Quantity")
plt.ylabel("Product Name");
plt.title('Which product has most number of sales?');


for patch in ax.patches:
    width = patch.get_width() 
    height = patch.get_height() 
    x = patch.get_x() 
    y = patch.get_y()
    
    plt.text(width + x, height + y, ' {:.1f} '.format(width))
```
